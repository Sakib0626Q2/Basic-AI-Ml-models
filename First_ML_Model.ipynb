```python
# prompt: create a gui of the above code and generate link of it with Predict fruit button

import pandas as pd
import numpy as np
from sklearn import tree
import gradio as gr

# AI is achieved through a process called ML
# apples and oranges, not in image but numerical form

# Apple1 = weight, texture, 160g, solid/0
# orange1 = 130, bumpy/1

# Apple2 = weight, texture, 170g, solid
# orange2 = 120, bumpy

# Apple1 = weight, texture, 180g, solid
# orange1 = 110, bumpy


features = [[160, 0], [130, 1], [170, 0], [120, 1], [180, 0], [110, 1]] #input
label = ["Apple", "Orange", "Apple", "Orange", "Apple", "Orange"] #output

df = pd.DataFrame(features, columns=["Height", "Surface"], index = label)

classifier = tree.DecisionTreeClassifier()
myModel = classifier.fit(features, label) #input, output

def predict_fruit(height, surface):
    # The model expects input as a list of lists
    prediction = myModel.predict([[height, surface]])
    return prediction[0]

# Create the Gradio interface
interface = gr.Interface(
    fn=predict_fruit,
    inputs=[
        gr.Slider(minimum=100, maximum=200, label="Height (g)"),
        gr.Radio(choices=[0, 1], label="Surface (0: Solid, 1: Bumpy)")
    ],
    outputs="text",
    title="Fruit Predictor",
    description="Enter the height and surface texture of a fruit to predict if it's an Apple or an Orange.",
    live=False,
    allow_flagging=False
)

# Launch the Gradio interface and generate a shareable link
interface.launch(share=True)
```
